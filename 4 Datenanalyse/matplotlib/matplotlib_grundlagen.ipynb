# Matplotlib Grundlagen
## Was ist Matplotlib?
Matplotlib ist ein umfassendes Modul in Python für die Erstellung von statischen, animierten und interaktiven Visualisierungen. Sie ist besonders nützlich für die Darstellung von Daten in Form von Diagrammen und Grafiken und wird häufig im Bereich Data Science und Machine Learning verwendet.</br>
Wir beschäftigen und hauptsächlich mit einer der wichtigsten Komponente: `pyplot`.

Installation via pip:
`pip install matplotlib`

```python
import matplotlib.pyplot as plt
```

Wir erstellen zunächst ein einfaches Diagramm.

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Liniendiagramm erstellen
plt.plot(x, y)

# Titel und Beschreibungen, wenn gewünscht (nicht notwendig, aber schöner)
plt.title("Einfaches Liniendiagramm")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")

# Anzeigen des Diagramms
plt.show()
```

### Weitere Plot-Typen

#### Balkendiagramm:
`plt.bar(x, y)`

```python
plt.bar(x, y)
plt.show()
```

#### Streudiagramm:
`plt.scatter(x, y)`

```python
plt.scatter(x, y)
plt.show()
```

#### Kuchendiagramm:
`plt.pie(sizes, labels=labels)`<br>
`sizes` = die Verteilung der Kuchenstücke<br>
`labels` = die Klasse/Beschreibung/Name des Kuchenstücks

```python
sizes = [0.6, 0.3, 0.2, 0.05, 0.05]
labels = ["Python ist cool!", "Python ist super!", "Python ist mega!", "Python ist okay.", "Python ist doof."]

plt.pie(sizes, labels=labels)
plt.show()
```

#### Histogramm:
`plt.hist(data, bins=10)`<br>
`data` = Liste (o.ä.) mit Werten<br>
`bins` = Einteilung in Klassen/"Eimer"

<br>
Nutzt intern Funktionen von numpy.

```python
data = [1, 1, 1, 2, 5, 4, 4, 5, 3, 2, 3, 4, 1, 2]

plt.hist(data, bins=5)
plt.show()
```

## Anpassen von Diagrammen
Matplotlib bietet viele verschiedene Möglichkeiten, Farben, Schriftgröße etc. zu ändern.

```python
# Farbe und Linienstil ändern
plt.plot(x, y, color='green', linestyle='dashed', marker='o')

# Titel und Achsenbeschriftung
plt.title("Titel des Diagramms")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")

# Legende hinzufügen, dazu brauchen wir labels, siehe später
plt.legend(loc="upper left")

plt.show()
```

## Subplots: Mehrere Diagramme in einer Darstellung
Wir können auch mehrere Diagramme in einer Ansicht erstellen. Dazu nutzen wir den Befehl `subplots`.

```python
# 2x2 Raster mit Diagrammen erstellen
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 1].bar(x, y)
axs[1, 0].scatter(x, y)
axs[1, 1].hist(y)
plt.show()
```

```python
# man kann aber auch mehrere Plots in ein und dem selben Diagramm erstellen
y1 = [i**2 for i in x]
y2 = [i/2 for i in x]

plt.plot(x, y1, label="quadriert") # label für die Legende
plt.scatter(x, y2, label="halbiert") # label für die Legende
plt.legend(loc="upper left")
plt.show()
```

## Speichern von Diagrammen

```python
plt.plot(x, y1, label="quadriert", linestyle='dashed', marker='o', color="red")
plt.legend()
plt.savefig("diagramm.png")
```

## Beispiel: Matplotlib + NumPy

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten erstellen
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Figur und der Achsen
fig, ax = plt.subplots()

# Hinzufügen der Daten zu den plots
ax.plot(x, y1, label="sin(x)")
ax.plot(x, y2, label="cos(x)", linestyle='--')

# Titel und Beschriftungen
ax.set_title("Sinus und Cosinus")
# bei Subplots nutzt man ax.set_xlabel statt plt.xlabel
ax.set_xlabel("X-Werte")
ax.set_ylabel("Y-Werte")

# Legende
ax.legend()

# Speichern
plt.savefig("sinus_cosinus.png")

# Anzeigen des Diagramms
plt.show()
```
